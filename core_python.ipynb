##Variables
Containers that stores data

##Types
What type of data is inside the container (number,text,True/False)

##Control Flow
Rules that decide what happens next


```python
# Store a person's name
name = "Maruthi"

# Store age
age = 21

# Store whether the person is a student
is_student = True

# Print values
print(name)
print(age)
print(is_student)

# Control flow using if statement
if age >= 18:
    print("Adult")
else:
    print("Minor")
```

##Functions
Functions takes input as parameters and do some process and retuen the output

##*args
Accept any number of items, args becomes a tuple

##**kwargs
Accept any number of labeled items, kwargs becomes a dictionary


```python
# Function to greet a user
def greet(name):
    # Print greeting message
    print("Hello", name)

# Call the function
greet("Maruthi")
```

```python
def add_numbers(*args):
    # args becomes a tuple
    print(args)

add_numbers(1, 2, 3)
```

```python
def user_info(**kwargs):
    # kwargs becomes a dictionary
    print(kwargs)

user_info(name="Maruthi", age=21)
```

##List and Dictionary Comprehensions

*Instead of writing many lines, you write one clean line.*



##List Comprehension
###Normal Way

```python
numbers = [1, 2, 3, 4]

squares = []

for n in numbers:
    squares.append(n * n)

print(squares)
```

###Shorter Way

```python
# Create square of each number
squares = [n * n for n in numbers]

print(squares)
```

##Dictionary Comprehension

```python
numbers = [1, 2, 3]

# Create dictionary
square_dict = {n: n * n for n in numbers}

print(square_dict)
```

##Generator Expressions
The code inside a function is not executed yet, it is only compiled. The function only executes when you iterate over the generator

```python
def my_generator():
  yield 1
  yield 2
  yield 3
  yield 4

for value in my_generator():
  print(value)
```

```python
numbers = (n*n for n in range(5))
# Get values one by one
print(next(numbers))
print(next(numbers))
```

##Type Hints
Type hints says, “This variable should store this kind of data.”

```python
# Function with type hints
def add(a: int, b: int) -> int:
    # Return sum
    return a + b

result = add(2, 3)

print(result)
```

##Decorators
A function that adds extra behaviour to another function without changing the original code

Before → Original Function → After

```python
# Decorator function
def decorator(func):

    # Wrapper function
    def wrapper():
        print("Before function")

        # Call original function
        func()

        print("After function")

    return wrapper

```

```python
# Apply decorator
@decorator
def say_hello():
    print("Hello!")

# Run function
say_hello()
```
