#What an LLM Actually Is

##Trained on a Fixed Snapshot

An LLM (Large Language Model) is trained using a huge collection of text from books, websites, articles, and conversations.

But it does not continuously learn after training.

It learns from a fixed snapshot of data collected at a certain time.







##Knowledge Cutoff Dates and What They Imply

The model only knows information before a certain date.

Example:

* If cutoff = January 2025
* The model may not know events after January 2025



##Probabilistic Generation, Not Retrieval

LLMs do not search memory like Google.

Instead:

* They predict the next most likely word
* One word at a time

This process is called:

Probabilistic generation

The model generates text based on patterns learned during training.

##Why the Same Prompt Gives Different Outputs

LLMs can generate different answers for the same question because:

* Multiple responses may be correct
* The model uses probabilities
* Some randomness is added during generation

This makes conversations feel natural.
