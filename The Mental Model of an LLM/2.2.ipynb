#How an LLM thinks

##BPE tokenization — why "hello" is 1 token but "antidisestablishmentarianism" is 6

LLMs do not read full sentences exactly like humans.

They first break text into smaller pieces called:

Tokens

A token can be:

* A word
* Part of a word
* A symbol
* A punctuation mark

BPE (Byte Pair Encoding) is one popular method used for tokenization.



##Context Windows — What Fits, What Gets Silently Truncated

An LLM can only remember a limited amount of text at one time.

This limit is called Context Window

If too much text is added:

* Older text may be removed
* Important information may be forgotten

This is called Truncation

##Transformer at 30,000 Feet — Attention Preserves Position

Transformers are the main technology behind LLMs.

A transformer uses Attention

Attention helps the model:

* Focus on important words
* Understand relationships between words
* Remember sentence meaning

##Why Long Context Degrades ("Lost in the Middle")

When conversations become too long:

* Models may forget middle information
* Beginning and ending parts are remembered better

This problem is called Lost in the Middle
