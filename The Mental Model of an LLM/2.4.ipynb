#Reading Model Evals & Benchmarks

##The Benchmarks Worth Knowing — MMLU, GSM8K, HumanEval, SWE-bench, GPQA, MMMU, BFCL

Benchmarks are tests used to measure how good an AI model is.

Different benchmarks test different skills.

These scores help developers compare AI models.

Screenshot from 2026-05-25 19-56-11.png

##Why Benchmarks Lie — Contamination, Prompt Sensitivity, Eval Gaming

Benchmark scores are not always fully trustworthy.

Sometimes:

* Models memorize answers
* Small prompt changes affect results
* Companies optimize only for benchmark scores

This can create misleading results.

##How to Read a Leaderboard Skeptically (LMArena, Artificial Analysis, Vellum, Aider)

Leaderboards rank AI models from best to worst.

But:

Higher score does not always mean better model

You should check:

8 What benchmark was used
* Real-world testing
* Speed
* Cost
* Reliability

Popular leaderboard platforms:

* LMArena
* Artificial Analysis
* Vellum
* Aiderh

## Building Your Own Micro-Eval for the Task You Actually Care About

A micro-eval is:

* A small custom test created for your own project

Instead of trusting public benchmarks,
you test models using your real tasks.

Example:

* Your chatbot questions
* Your coding tasks
* Your customer support prompts
