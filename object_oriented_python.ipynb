#Object Oriented Python
## classes,__init__,instance vs class methods

Object oriented programming helps us organizing code using classes and objects


*   A class is like a blueprint/template
*   An object is the real thing created from that blueprint


**Ex**:

  class -> Car blueprint

  object -> Actual car


__init__ is a special method that automatically runs when we create an object


**instance method** : work with object data

**class method** : work with class level data

###Memory tricks

self -> current object

cls -> current class

__init__ runs automatically

class variables are shared by all objects


```python
class Mobile:

    # Class variable
    brand = "Samsung"

    def __init__(self, model):
        self.model = model

    # Instance method
    def show_model(self):
        print("Model:", self.model)

    # Class method
    @classmethod
    def show_brand(cls):
        print("Brand:", cls.brand)

# Create object
m1 = Mobile("Galaxy S24")

# Call methods
m1.show_model()
Mobile.show_brand()
```

##Inheritance, Encapsulation, Polymorphism

###Inheritance

Inheritance allows one class to use properties and methods of another class.

This helps avoid rewriting code.

Child inherits features from parents.



```python
# Parent class
class Animal:

    def sound(self):
        print("Animal makes sound")

# Child class
class Dog(Animal):

    def bark(self):
        print("Dog barks")

# Create object
d = Dog()

# Access parent method
d.sound()

# Access child method
d.bark()
```

###Encapsulation

Encapsulation means hiding internal details and protecting data.

Double underscore (__) makes variable private.

```python
class Employee:

    def __init__(self):
        # Private variable
        self.__salary = 50000

    def show_salary(self):
        print("Salary:", self.__salary)

e = Employee()

# Access using method
e.show_salary()
```

###Polymorphism

Polymorphism means:

Same method name behaves differently.

**Ex:**

A person can:

* work as teacher
* act as parent
* drive as driver

Same person → different behavior.



```python
class Cat:

    def sound(self):
        print("Cat says Meow")

class Dog:

    def sound(self):
        print("Dog says Bark")

# Function using polymorphism
def make_sound(animal):
    animal.sound()

# Create objects
c = Cat()
d = Dog()

# Call function
make_sound(c)
make_sound(d)
```

**NOTE:**

* Inheritance → reuse code
* Encapsulation → protect data
* Polymorphism → same method, different behavior

###Dataclasses

Dataclasses automatically create:

* constructor
* print methods
* comparison methods

This reduces extra code.

```python
# Import dataclass
from dataclasses import dataclass

# Create dataclass
@dataclass
class Student:
    name: str
    age: int

# Create object
s1 = Student("Maruthi", 19)

# Print object
print(s1)
```

###Pydantic Models

Pydantic is used for:

* data validation
* checking correct data types
* creating structured data models

✅ Automatically validates data

✅ Prevents errors early

```python
# Import BaseModel
from pydantic import BaseModel

# Create model
class User(BaseModel):

    name: str
    age: int

# Create object
user1 = User(name="Maruthi", age=19)

# Print object
print(user1)
```
