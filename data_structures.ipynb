#Data Structures

##List

A List in Python is used to store multiple items in one variable.

Lists are:

* Ordered
* Changeable (mutable)
* Allow duplicate values


Screenshot from 2026-05-20 11-23-03.png

```python
# Creating a list
fruits = ["Apple", "Banana", "Mango"]

# Printing the entire list
print(fruits)

# Accessing first item
print(fruits[0])

# Adding a new item
fruits.append("Orange")

# Printing updated list
print(fruits)
```

##Tuple

A Tuple is similar to a list, but it cannot be changed after creation.

Tuples are:

* Ordered
* Immutable (unchangeable)
* Allow duplicates



Screenshot from 2026-05-20 11-27-24.png

```python
# Creating a tuple
colors = ("Red", "Green", "Blue")

# Printing tuple
print(colors)

# Accessing item
print(colors[1])
```

##Set

A Set stores unique values only.

Sets:
* do not allow duplicates
* are unordered
* are mutable



Screenshot from 2026-05-20 11-30-17.png

```python
# Creating a set
numbers = {1, 2, 2, 3, 4}

# Printing set
print(numbers)

# Adding a value
numbers.add(5)

# Printing updated set
print(numbers)
```

##Dictionary

A Dictionary stores data in:

key : value

Screenshot from 2026-05-20 11-31-10.png

```python
# Creating dictionary
student = {
    "name": "John",
    "age": 21
}

# Printing dictionary
print(student)

# Accessing value using key
print(student["name"])

# Adding new key-value pair
student["city"] = "Delhi"

print(student)
```

##NamedTuple

NamedTuple is like a tuple, but with named fields.

It makes code easier to read.

```python
# Importing namedtuple
from collections import namedtuple

# Creating structure
Student = namedtuple("Student", ["name", "age"])

# Creating object
s1 = Student("John", 21)

# Accessing values
print(s1.name)
print(s1.age)
```

##defaultdict

defaultdict automatically creates default values for missing keys.



Screenshot from 2026-05-20 11-39-47.png

```python
from collections import defaultdict

# Default value is 0
count = defaultdict(int)

# Incrementing values
count["apple"] += 1
count["apple"] += 1

print(count["apple"])
```

## Counter

Counter counts how many times items appear.

Screenshot from 2026-05-20 11-39-57.png

```python
from collections import Counter

# List of items
fruits = ["apple", "banana", "apple"]

# Counting items
count = Counter(fruits)

print(count)
```

## deque

deque is a fast list for adding/removing items from both ends.

Screenshot from 2026-05-20 11-40-10.png

```python
from collections import deque

# Creating deque
queue = deque([1, 2, 3])

# Add item at end
queue.append(4)

# Remove from front
queue.popleft()

print(queue)
```
