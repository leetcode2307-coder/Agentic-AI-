##First /chat Endpoint

An endpoint is a URL in your API that performs a specific task.

In FastAPI, we create endpoints so users or apps can send requests and receive responses.

Example:

* /chat
* /login
* /weather

The /chat endpoint can receive a message and return a reply.

```python
# Import FastAPI
from fastapi import FastAPI

# Create app object
app = FastAPI()

# Create endpoint
@app.get("/chat")
def chat():

    # Return response
    return {"reply": "Hello User!"}
```

http://127.0.0.1:8000/chat

## Pydantic Request/Response Models

Pydantic models help us define:

* What data should come into API
* What data should go out

It validates data automatically.

```python
from fastapi import FastAPI
from pydantic import BaseModel

# Create app
app = FastAPI()

# Create request model
class ChatRequest(BaseModel):

    # User message
    message: str

# Create POST endpoint
@app.post("/chat")
def chat(data: ChatRequest):

    # Return user message
    return {"you_said": data.message}
```

## Dependency Injection

Dependency Injection means:
“Automatically provide something a function needs.”

FastAPI can automatically give:

* Database connection
* User authentication
* API keys
Shared functions

```python
from fastapi import FastAPI, Depends

# Create app
app = FastAPI()

# Dependency function
def get_user():

    # Return username
    return "Maruthi"

# Endpoint
@app.get("/")
def home(user: str = Depends(get_user)):

    # Return dependency value
    return {"user": user}
```

## Automatic OpenAPI Docs

## Running with Uvicorn

Uvicorn is the server that runs FastAPI applications.

Without Uvicorn, FastAPI code will not start.



```python
uvicorn main:app --reload
```

main -> Python filename (main.py)

app -> FastAPI object name.

--reload -> Automatically restarts server after code changes.
