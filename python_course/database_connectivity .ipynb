#Database Connectivity

##psycopg2 for raw PostgreSQL

psycopg2 is a Python library used to connect Python programs with a PostgreSQL database.

It allows Python to:

Send SQL queries,
Insert data,
Read data,
Update records,
Delete records

Think of it as a bridge between Python and PostgreSQL.



```python
import psycopg2

# Connect to PostgreSQL database
conn = psycopg2.connect(
    host="localhost",
    database="school",
    user="postgres",
    password="1234"
)

# Create cursor object
cur = conn.cursor()

# Run SQL query
cur.execute("SELECT 'Hello Database'")

# Get result
result = cur.fetchone()

# Print result
print(result)

# Close connection
cur.close()
conn.close()
```

Line-by-line
* import psycopg2 → Imports database library
* connect() → Connects Python to PostgreSQL
* cursor() → Used to run SQL commands
* execute() → Runs SQL query
* fetchall() → Gets all rows
* close() → Closes connection safely

##SQLAlchemy ORM Basics

SQLAlchemy ORM allows us to work with databases using Python classes instead of writing raw SQL queries.

ORM means Object Relational Mapping




```python
from sqlalchemy import create_engine, Column, Integer, String
from sqlalchemy.orm import declarative_base

# Create database engine
engine = create_engine("sqlite:///school.db")

# Create base class
Base = declarative_base()

# Create table model
class Student(Base):
    __tablename__ = "students"

    # Table columns
    id = Column(Integer, primary_key=True)
    name = Column(String)

# Create tables inside database
Base.metadata.create_all(engine)

print("Table created successfully")
```

##Connection Pooling and Why It Matters Under Load

Connection pooling means:

Reusing database connections instead of creating new ones every time.

Creating database connections is expensive and slow.

Pooling saves time and improves performance.

```python
from sqlalchemy import create_engine

# Create engine with connection pool
engine = create_engine(
    "sqlite:///company.db",
    pool_size=5
)

print("Connection pool created")
```

Line-by-line
* pool_size=5 → Keeps 5 reusable connections ready
* Reusing connections improves speed

##Raw SQL When the ORM Gets in the Way

Sometimes ORM is not enough.

Complex queries can become:

* Slow
* Difficult
* Confusing

In such cases, developers directly write SQL queries.

This is called Raw SQL.

```python
from sqlalchemy import create_engine, text

# Create database connection
engine = create_engine("sqlite:///school.db")

# Open database connection
with engine.connect() as conn:

    # Execute raw SQL query
    result = conn.execute(
        text("SELECT 'Raw SQL Working'")
    )

    # Print query result
    for row in result:
        print(row)
```
