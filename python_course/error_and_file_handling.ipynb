#Error & File Handling

##try / except / finally

Sometimes our program crashes because of mistakes like:

* dividing by zero
* opening a missing file
* wrong user input

Python gives errors (exceptions) when something goes wrong.

We use:

try → code that may cause an error

except → handles the error safely

finally → runs no matter what

This prevents the program from suddenly stopping.

```python
try:
    # Asking user for a number
    number = int(input("Enter a number: "))

    # Division operation
    result = 10 / number

    print("Result:", result)

except ZeroDivisionError:
    # Runs if number is 0
    print("You cannot divide by zero!")

except ValueError:
    # Runs if input is not a number
    print("Please enter a valid number.")

finally:
    # Always runs
    print("Program finished.")
```

## Custom Exception Classes

Python allows us to create our own errors.

These are called custom exceptions.

We create them when built-in errors are not enough.


```python
# Creating custom exception
class AgeError(Exception):
    pass


age = int(input("Enter your age: "))

try:
    if age < 18:
        # Raising custom error
        raise AgeError("You must be 18 or older.")

    print("Access granted.")

except AgeError as e:
    print("Error:", e)
```

## Context Managers (with, contextlib)

A context manager helps manage resources automatically.

Example:

opening files

database connections

network connections

The with statement automatically cleans things up.

```python
# Writing into a file
with open("notes.txt", "w") as file:

    # Writing text
    file.write("Hello Python")

print("File written successfully.")
```

###Using contextlib

contextlib helps create custom context managers.



```python
from contextlib import contextmanager


@contextmanager
def my_context():

    print("Start")

    yield

    print("End")


with my_context():
    print("Inside block")
```

##Reading/Writing JSON, CSV, Plain Text, Binary Files

###Plain Text Files

Text files store normal readable text.



```python
# Writing text file
with open("message.txt", "w") as file:
    file.write("Hello World")

# Reading text file
with open("message.txt", "r") as file:
    data = file.read()

print(data)
```

###JSON Files

JSON stores data in key-value format.

Looks similar to Python dictionaries.


```python
import json

student = {
    "name": "John",
    "age": 20
}

# Writing JSON file
with open("student.json", "w") as file:
    json.dump(student, file)

# Reading JSON file
with open("student.json", "r") as file:
    data = json.load(file)

print(data)
```

###CSV Files

CSV files store table data.

CSV = Comma Separated Values



```python
import csv

# Writing CSV file
with open("students.csv", "w", newline="") as file:

    writer = csv.writer(file)

    writer.writerow(["Name", "Age"])
    writer.writerow(["Alice", 20])

# Reading CSV file
with open("students.csv", "r") as file:

    reader = csv.reader(file)

    for row in reader:
        print(row)
```

###Binary Files

Binary files store data in machine-readable format.

Examples:

* images
* videos
* PDFs

```python
# Writing binary file
with open("data.bin", "wb") as file:

    file.write(b"Hello")

# Reading binary file
with open("data.bin", "rb") as file:

    data = file.read()

print(data)
```
