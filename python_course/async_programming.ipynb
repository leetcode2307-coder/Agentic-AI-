#Async Programming

##asyncio fundamentals — event loop, coroutines

Imagine a restaurant waiter:

* The waiter takes one order.
* While food is cooking, the waiter does NOT stand idle.
* Instead, the waiter serves other tables.

This is how async programming works.

Python can start a task, wait for it, and meanwhile work on other tasks.

###Coroutine
A coroutine is a special function that can “pause” and “resume”.

You create it using :
async def

###Event Loop

The event loop is like the restaurant manager.

It:

* keeps track of tasks
* switches between them
* runs tasks when they are ready




```python
import asyncio

# Create an async function (coroutine)
async def make_tea():

    # Print starting message
    print("Boiling water...")

    # Wait for 2 seconds without blocking program
    await asyncio.sleep(2)

    # Continue after waiting
    print("Tea is ready!")

# Start the event loop and run coroutine
await make_tea()
```

##async/await syntax

Think of ordering food online.

* async = “This task may take time”
* await = “Wait here until result comes



```python
import asyncio

# Async function to download file
async def download_file():

    print("Downloading started...")

    # Simulate internet delay
    await asyncio.sleep(3)

    print("Download completed!")

# Run program
await download_file()
```

##asyncio.gather() for Parallel LLM Calls

Suppose you ask:

* ChatGPT
* Gemini
* Claude

the same question at the same time.

Instead of waiting one-by-one, you ask all together.



```python
import asyncio

# First async task
async def task1():

    print("Task 1 started")

    # Simulate waiting
    await asyncio.sleep(2)

    print("Task 1 completed")

# Second async task
async def task2():

    print("Task 2 started")

    # Simulate waiting
    await asyncio.sleep(1)

    print("Task 2 completed")

# Main async function
async def main():

    # Run both tasks together
    await asyncio.gather(
        task1(),
        task2()
    )

# Run event loop
await main()
```

##asyncio.wait_for() for Timeout Protection

Imagine waiting for a friend.

You say:

    “I’ll wait only 5 minutes.”

If they don’t come, you leave.

That’s timeout protection.



```python
import asyncio

# Slow async task
async def slow_task():

    # Wait for 5 seconds
    await asyncio.sleep(5)

    return "Finished"

async def main():

    try:

        # Wait only 2 seconds
        result = await asyncio.wait_for(
            slow_task(),
            timeout=2
        )

        print(result)

    except asyncio.TimeoutError:

        print("Task took too long!")

# Run program
await main()
```

## asyncio.create_task() for Fire-and-Forget Logging

Imagine sending a background worker to:

* write logs
* save analytics
* send notifications

You continue your main work without waiting.

That is:

`asyncio.create_task()`



```python
import asyncio

# Background logging task
async def save_log():

    # Simulate file saving
    await asyncio.sleep(2)

    print("Log saved")

# Main task
async def main():

    print("Main task started")

    # Start background task
    asyncio.create_task(save_log())

    print("Main task continues")

    # Wait so background task can finish
    await asyncio.sleep(3)

# Run program
await main()
```
